```python
rates = pd.DataFrame({"euribor": euribor.rate, "sp500": sp500["Long Interest Rate"]})
rates = rates.resample("1d").interpolate()
rates = rates.dropna()
rates.plot()
```

```python
rates = pd.DataFrame({
    "euribor": euribor["rate"],
    "sp500": sp500["Long Interest Rate"]
})
rates = rates.resample('1d').interpolate()
#print(rates)
rates = rates.dropna()
rates.plot()
```

```python
iris = pd.read_csv(
    "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
    header=None,
    names=["sepal_length", "sepal_width", "petal_length",
           "petal_width", "name"])
iris.groupby(iris.name).mean()
```

```python
# 1. Filtern: nur Wechselkurse aus den 90ern
# 2. Gruppieren nach Land
# 3. Basierend auf den Gruppen Durchschnitte berechnen

exchange_rates_90s = exchange_rates[
    (exchange_rates["Date"] >= "1990-01-01") &
    (exchange_rates["Date"] < "2000-01-01")
]
er_90s_grouped = exchange_rates_90s.groupby("Country")
er_90s_grouped.mean()
```
