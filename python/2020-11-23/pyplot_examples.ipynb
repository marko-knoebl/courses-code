```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn")

x = np.linspace(0, 2*np.pi, 100)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x, sin, "C0--", label="sin(x)")
plt.plot(x, cos, "C1:", label="cos(x)")

pi_multiples = np.array([0, 0.5, 1, 1.5, 2]) * np.pi
sin_points = np.sin(pi_multiples)
cos_points = np.cos(pi_multiples)
plt.plot(pi_multiples, np.sin(pi_multiples), "C0o")
plt.plot(pi_multiples, cos_points, "C1o")

plt.title("Trigonometric functions")
plt.xlabel("x (radians)")
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.legend()
plt.axis("scaled")
```

```python
def plot_gaussian_function(mu, sigma, color="C0", linestyle="dashed"):
    """Plot a gaussian bell curve

    mu = mean value
    sigma = standard deviation
    """
    # plot between mu-4*sigma and mu+4*sigma

    x = np.linspace(mu-4*sigma, mu+4*sigma, 200)
    y = 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-0.5 * (x - mu)**2 / sigma**2)

    plt.plot(x, y, color=color, linestyle=linestyle)

plot_gaussian_function(0, 1)
plot_gaussian_function(1, 1, color="C1", linestyle="solid")
plot_gaussian_function(0, 2, color="C2")
plot_gaussian_function(0, 3, "C3", "solid")
plot_gaussian_function(0, 4, linestyle="solid", color="C4")
plot_gaussian_function(0, 5, linestyle="dotted")
```

```python
# dice rolls

# 3, 2, 4, 1, 6, 6, 5, ... -> 37
# 4, 3, 6, 6, 1,           -> 40
#                             34
#                             35

simulation_10_dice = np.random.randint(1, 7, (10, ) )
sum_10_dice = np.sum(simulation_10_dice)
print(sum_10_dice)

simulation_1_million_times_10_dice = np.random.randint(1, 7, (1_000_000, 10))

simulation_1_million_times_10_dice[:5, :]

sums_1_million_times_10_dice = np.sum(simulation_1_million_times_10_dice, axis=1)
print(sums_1_million_times_10_dice)

plt.hist(
    sums_1_million_times_10_dice,
    bins=[15, 20, 25, 30, 35, 40, 45, 50, 55]
)
```

```python
import pandas as pd
iris = pd.read_csv(
    "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
)
# get all rows and the first four columns as numpy data
iris = iris.iloc[:,:4].to_numpy()

sepal_length = iris[:,0]
sepal_width = iris[:,1]
petal_length = iris[:,2]
petal_width = iris[:,3]
```

```python
plt.plot(petal_length[:50], petal_width[:50], "o", label="setosa")
plt.plot(petal_length[50:100], petal_width[50:100], "o", label="versicolor")
plt.plot(petal_length[100:150], petal_width[100:150], "o", label="virginica")

plt.legend()
```

```python
plt.scatter(petal_length, petal_width, s=sepal_length**3, c=sepal_width)
plt.colorbar()
```

```python
plt.hist(
    petal_length,
    bins=np.arange(0.5, 7.5, 0.5),
    rwidth=0.8
)
```

```python
fig = plt.figure(
    figsize=(4, 6),
    facecolor="#eeeeee"
)
plt.boxplot(
    [petal_length, petal_width, sepal_length, sepal_width],
    labels=["petal length", "petal width", "sepal length",
            "sepal width"]
)
fig.savefig("iris-boxplots.png")
```
