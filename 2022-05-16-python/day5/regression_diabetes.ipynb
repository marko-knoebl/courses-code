```python
from sklearn.datasets import load_diabetes

# load dataset as a pandas dataframe
dataset = load_diabetes(as_frame=True)
```

```python
dataset.target
```

```python
import statsmodels.api as sm

x = sm.add_constant(dataset.data)
```

```python
x
```

```python
y = dataset.target
```

```python
model = sm.OLS(y, x)

res = model.fit()

res.summary()
```

```python
import pandas as pd

x_0 = pd.Series([1, 1, 1, 1, 1])

x_1 = pd.Series([1, 2, 3, 4, 5])

x_2 = x_1**2
x_3 = x_1**3

x = pd.DataFrame({"x_0": x_0, "x_1": x_1, "x_2": x_2, "x_3": x_3})

y = pd.Series([2, 4, 9, 15, 31])
```

```python
model = sm.OLS(y, x)

res = model.fit()

res.summary()
```
