# Loading Data

```python
import pandas as pd
```

```python
data = pd.read_csv("../data/penguins.csv")
data
```

# Filtering / Querying Data

```python
# this code is not clean / ideal!
# good code would do this in just one line via isna() / dropna()

na_list = []

for index, row in data.iterrows():
    if pd.isna(row["bill_length_mm"]) or pd.isna(row["bill_depth_mm"]) or pd.isna(row["flipper_length_mm"]) or pd.isna(row["body_mass_g"]):
        na_list.append(row)

pd.DataFrame(na_list)
```

```python
# this code is not clean / ideal!
# good code would do this in just one line via isna() / dropna()

na_list = []

for index, row in data.iterrows():
    if pd.isna(row["sex"]):
        na_list.append(row)

pd.DataFrame(na_list)
```
