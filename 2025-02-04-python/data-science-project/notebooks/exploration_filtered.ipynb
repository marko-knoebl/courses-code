# Data processing and visualization

```python
data = pd.read_csv("../data/penguins.csv")
```

```python
valid_df = filter_out_nas(data)
```

```python
# replace strings with numeric values
valid_df["numeric_species"] = valid_df["species"].replace({"Adelie": 3, "Gentoo": 1, "Chinstrap": 2})

# create a colormap with 3 entries
cmap = plt.get_cmap("viridis", 3)

valid_df.plot.scatter(
    x="bill_length_mm",
    y="bill_depth_mm",
    c="numeric_species",
    cmap=cmap
)
```
