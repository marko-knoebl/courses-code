```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3])

y1 = x*2
y2 = x**2

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, 10 - x)
```

```python
# plot of sin and cos

# [0, 0.1, 0.2, ...]
x = np.arange(0, 2*np.pi, 0.1)

sin_x = np.sin(x)
cos_x = np.cos(x)

plt.plot(x, sin_x, color="k", linestyle="dashed")
plt.plot(x, cos_x, marker="X", color=(1, 0.7, 0));
```

```python
def plot_gaussian(mu, sigma):
    x = np.arange(-5, 5, 0.1)

    y = (
        1 / (sigma * np.sqrt(2 * np.pi)) *
          np.exp(- 1/2 * (x-mu)**2 / sigma**2)
    )

    plt.plot(x, y)

plt.style.use("default")

plot_gaussian(0, 1)
plot_gaussian(1, 2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn")

x = np.linspace(0, 2*np.pi, 100)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x, sin, "C0--", label="sin(x)")
plt.plot(x, -cos, "C1:", label="cos(x)")

pi_multiples = np.array([0, 0.5, 1, 1.5, 2]) * np.pi
sin_points = np.sin(pi_multiples)
cos_points = np.cos(pi_multiples)
plt.plot(pi_multiples, sin_points, "C0o")
plt.plot(pi_multiples, -cos_points, "C1o")

plt.title("Trigonometric functions")
plt.xlabel("x (radians)")
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.legend()
plt.axis("scaled")
```
