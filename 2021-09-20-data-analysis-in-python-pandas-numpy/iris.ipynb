```python
import pandas as pd

iris = pd.read_csv(
    "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
    header=None,
    names=["sepal_length", "sepal_width", "petal_length",
           "petal_width", "species"],
)
```

```python
iris[["sepal_length", "sepal_width"]].median().plot.bar()
```

```python
iris.plot.scatter(
    x="sepal_length",
    y="sepal_width",
    s="petal_length",
    c="petal_width"
)
```

```python
iris.plot.box()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(iris);
```
