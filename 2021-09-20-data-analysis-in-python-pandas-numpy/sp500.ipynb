```python
import pandas as pd

sp500 = pd.read_csv(
    "https://datahub.io/core/s-and-p-500/r/data.csv",
    index_col="Date",
    parse_dates=["Date"],
)
```

```python
sp500
```

```python
sp500["diff"] = sp500["SP500"].diff()
```

```python
sp500["diff"]
```

```python
sp500["gain"] = sp500["diff"] / sp500["SP500"]
```

```python
sp500
```

```python
sp500["gain"].loc["2016":].plot.bar()
```
