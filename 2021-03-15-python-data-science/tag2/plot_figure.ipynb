```python
import matplotlib.pyplot as plt

fig = plt.figure(
    figsize=(8, 6),
    facecolor="#eeeeee"
)
ax = plt.gca()

ax.plot([0, 1, 2, 3], [0, 1, 4, 9])
fig.savefig("test.png")
```
