```python
def pareto_diagram(data):
    data.sort_values(ascending=False).plot.bar()
    data.sort_values(ascending=False).cumsum().plot.line(color="C1")
```

```python
pareto_diagram(countries["pop"])
```

```python
pareto_diagram(countries["ar"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3])

y1 = x*2
y2 = x**2

plt.style.use("seaborn")

plt.plot(x, y1, color=(1, 0.7, 0))
plt.plot(x, y2, color="C0", marker="s", linestyle=":")
```

```python
x = np.linspace(0, 6.3, 100)

plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, -np.cos(x), label='cos(x)')

plt.legend()
```
