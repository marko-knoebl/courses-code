```python
prices = np.array([3.99, 4.99, 3.99, 12.99])
# buying the first item 3 times and the last item 2 times
quantities = np.array([3, 0, 0, 2])
```

```python
total_price_per_product = prices * quantities
total_price_per_product
```

```python
np.sum(total_price_per_product)
```

```python
def sumx(a, b):
    return np.sum(a * b)
```

```python
sumx(prices, quantities)
```
