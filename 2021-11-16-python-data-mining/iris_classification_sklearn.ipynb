```python
import pandas as pd
```

```python
# load data via pandas
iris = pd.read_csv(
    "http://archive.ics.uci.edu/ml/" +
    "machine-learning-databases/iris/iris.data",
    header=None,
    names=["sepal_length", "sepal_width", "petal_length",
           "petal_width", "species"]
)
```

```python
iris
```

```python
iris_shuffled = iris.sample(frac=1.0)
iris_shuffled
```

```python
measurements = iris_shuffled[[
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
]].to_numpy()
```

```python
species = (
    iris_shuffled["species"]
    .replace({
        "Iris-setosa": 0,
        "Iris-versicolor": 1,
        "Iris-virginica": 2,
    })
    .to_numpy()
)
```

```python
species
```

```python
x = measurements
```

```python
y = species
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
```

```python
model.fit(x[:130], y[:130])
```

```python
import numpy as np
demo_measurements = np.array([
    [5.3, 3.4, 1.9, 0.6],
    [6.0, 3.0, 4.7, 1.5],
    [6.5, 3.1, 5.0, 1.7]
])
```

```python
model.predict(demo_measurements)
```

```python
model.predict_proba(demo_measurements)
```

```python
measurements_val = measurements[130:]
species_val = species[130:]
```

```python
model.score(measurements_val, species_val)
```
